```python
import itertools
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Step 1: Define probabilities and odds for each match
matches = [
    {'team': 'Legia Warszawa vs Brøndby IF', 'win': 0.47, 'draw': 0.29, 'loss': 0.30, 'win_odds': 2.10, 'draw_odds': 3.40, 'loss_odds': 3.30},
    {'team': 'FC Pyunik Yerevan vs FC Ordabasy Shymkent', 'win': 0.33, 'draw': 0.31, 'loss': 0.40, 'win_odds': 3.00, 'draw_odds': 3.20, 'loss_odds': 2.50},
    {'team': 'FC Flora Tallinn vs Vikingur Reykjavik', 'win': 0.32, 'draw': 0.31, 'loss': 0.43, 'win_odds': 3.10, 'draw_odds': 3.20, 'loss_odds': 2.30},
    {'team': 'Sabah Masazir vs St Patrick\'s Athletic', 'win': 0.29, 'draw': 0.30, 'loss': 0.45, 'win_odds': 3.40, 'draw_odds': 3.30, 'loss_odds': 2.20},
    {'team': 'Zira FK vs NK Osijek', 'win': 0.30, 'draw': 0.29, 'loss': 0.47, 'win_odds': 3.30, 'draw_odds': 3.40, 'loss_odds': 2.10},
    {'team': 'Lincoln Red Imps vs FC Dinamo Minsk', 'win': 0.28, 'draw': 0.30, 'loss': 0.50, 'win_odds': 3.60, 'draw_odds': 3.30, 'loss_odds': 2.00},
    {'team': 'Ludmannsdorf vs SC St. Stefan', 'win': 0.32, 'draw': 0.29, 'loss': 0.40, 'win_odds': 3.10, 'draw_odds': 3.40, 'loss_odds': 2.50},
]
```

```python
# Step 2: Generate all possible combinations of outcomes (win, draw, loss)
outcomes = ['win', 'draw', 'loss']
combinations = list(itertools.product(outcomes, repeat=len(matches)))
```

```python
# Step 3: Calculate total probability, total odds, and expected value for each combination
results = []
for combo in combinations:
    total_prob = 1.0
    total_odds = 1.0
    for i, outcome in enumerate(combo):
        match = matches[i]
        total_prob *= match[outcome]
        total_odds *= match[f'{outcome}_odds']
    # Calculate the expected value for the combination
    expected_value = (total_prob * total_odds) - (1 - total_prob)
    results.append((combo, total_prob, total_odds, expected_value))
```

```python
# Assuming 'combinations' is a list of tuples where each tuple represents a combination of outcomes (e.g., ('win', 'draw', 'loss'))
from itertools import product

# Generate all possible combinations of 'win', 'draw', 'loss' for each match
outcome_types = ['win', 'draw', 'loss']
combinations = list(product(outcome_types, repeat=len(matches)))

# Step 3: Calculate total probability, total odds, and expected value for each combination
results = []
for combo in combinations:
    total_prob = 1.0
    total_odds = 1.0
    match_labels = []
    
    for i, outcome in enumerate(combo):
        match = matches[i]
        total_prob *= match[outcome]
        total_odds *= match[f'{outcome}_odds']
        match_labels.append(f"{match['team']} ({outcome})")
    
    # Calculate the expected value for the combination
    expected_value = (total_prob * total_odds) - (1 - total_prob)
    
    # Append the results with match labels
    results.append((*match_labels, total_prob, total_odds, expected_value))

# Convert results to a DataFrame for easier handling and viewing
columns = [f'Match {i+1}' for i in range(len(matches))] + ['Total Probability', 'Total Odds', 'Expected Value']
df_results = pd.DataFrame(results, columns=columns)

# Display the results
df_results.head()
```

```python
# Step 5: Filter combinations with a positive expected value
positive_ev_bets = df_results[df_results['Expected Value'] > df_results['Expected Value'].quantile(0.50) ]
positive_ev_bets
```

```python
# Step 6: Sort by Expected Value, then by Total Probability, then by Total Odds
sorted_bets = positive_ev_bets.sort_values(by=['Total Probability'], ascending=[False])
```

```python
# Step 6: Sort by Expected Value, then by Total Probability, then by Total Odds
sorted_bets = positive_ev_bets.sort_values(by=['Expected Value', 'Total Probability', 'Total Odds'], ascending=[False, False, False])
sorted_bets
```

```python
# Step 7: Select the top 10 bets (adjust the number based on your preference)
top_bets = sorted_bets.head(10)
top_bets
```

```python
# Step 8: Visualize the relationship between expected value, probability, and odds
plt.figure(figsize=(10, 6))
plt.scatter(sorted_bets['Total Probability'], sorted_bets['Expected Value'], c=sorted_bets['Total Odds'], cmap='viridis', s=50)
plt.colorbar(label='Total Odds')
plt.xlabel('Total Probability')
plt.ylabel('Expected Value')
plt.title('Expected Value vs Probability (Color indicates Total Odds)')
plt.show()
```

```python
import numpy as np

# Define thresholds for high expected value, reasonable probability, and balanced odds
expected_value_threshold = df_results['Expected Value'].quantile(0.85)  # Top 25% for expected value
probability_threshold_min = df_results['Total Probability'].quantile(0.45)  # Above the 25th percentile for probability
probability_threshold_max = df_results['Total Probability'].quantile(0.90)  # Below the 75th percentile for probability
odds_threshold_min = df_results['Total Odds'].quantile(0.25)  # Avoid very high odds, which might be too risky

# Apply filtering based on the criteria
filtered_bets = df_results[
    (df_results['Expected Value'] >= expected_value_threshold) &
    (df_results['Total Probability'] >= probability_threshold_min) &
    (df_results['Total Probability'] <= probability_threshold_max) &
    (df_results['Total Odds'] >= odds_threshold_min)
]

# Sort by Expected Value in descending order to get the best bets at the top
best_bets = filtered_bets.sort_values(by='Expected Value', ascending=False)

# Display the top 10 best bets based on these criteria
top_bets_based_on_criteria = best_bets.head(10)
print(top_bets_based_on_criteria)

# Optionally, you could visualize these on the graph
plt.figure(figsize=(10, 6))
plt.scatter(df_results['Total Probability'], df_results['Expected Value'], c=df_results['Total Odds'], cmap='viridis', s=50)
plt.colorbar(label='Total Odds')
plt.xlabel('Total Probability')
plt.ylabel('Expected Value')
plt.title('Expected Value vs Probability (Best Bets Highlighted)')
plt.scatter(
    top_bets_based_on_criteria['Total Probability'],
    top_bets_based_on_criteria['Expected Value'],
    color='red', s=100, label='Best Bets'
)
plt.scatter(
    top_bets['Total Probability'],
    top_bets['Expected Value'],
    color='pink', s=100, label='Top Bets'
)
plt.legend()
plt.show()
```

```python
# Step 10: Filter combinations to match actual results
filtered_combinations = []
matching_row_indices = []

for idx, (combo, prob, odds, ev) in enumerate(sorted_bets.itertuples(index=False)):
    matches_actual_results = True
    for i, outcome in enumerate(combo):
        match_name = matches[i]['team']
        if outcome != actual_results[match_name]:
            matches_actual_results = False
            break
    
    if matches_actual_results:
        filtered_combinations.append((combo, prob, odds, ev))
        matching_row_indices.append(idx)
```
